```python
import string
import secrets
```

```python
# obviously, this needs a lot more work, but I mainly made it to generate my UserID
class SKSessionHandler:
    userid_size = 3
    sessionid_size = 12

    def userIDGenerator():
        alphanumerics = string.ascii_uppercase + string.digits
        return ''.join(secrets.choice(alphanumerics) for i in range(SKSessionHandler.userid_size))

    def sessionIDGenerator():
        alphabet = string.ascii_letters + string.digits
        return ''.join(secrets.choice(alphabet) for i in range(SKSessionHandler.sessionid_size))
```

```python
print(SKSessionHandler.userIDGenerator())
```
